**[+1]** *В чём отличие между leaf-wise tree и depth-wise tree?*

```python
"""
Leaf-wise и depth-wise - это методы построения деревьев. Если используется leaf-wise построение дерева, то деревья строятся
по листьям. Лист выбирается тот при, котором потери сводятся к минимуму.
В depth-wise деревьях построение происходит по уровням и каждый узел разделяет данные и
дает приоритет тем узлам, которые ближе к корню.
Эти метода подходят разным данным, а именно если данных не много, то лучше использовать depth-wise tree и иначе,
если данных давольно много, то лучше использовать leaf-wise. Потому что lead-wise может переобучиться на малом
количесве данных. Так же leaf-wise tree лучше подходит большим выборкам потому что он строит деревья бысрее.
"""
```

**[+1]** *Как считается feature importance в выбранном вами бустинге?*

```python
"""
По умолчанию в Catboost важность признаков вычисляется методом PredictionValuesChange. 
Для каждого признака вычисляется, насколько в среднем прогноз изменяется при изменении значения признака.
То есть значение признака будет больше, если при изменении значения признака
в среднем больше изменилось значния предсказания.
"""
```

```python
"""
Пары листьев, которые сравниваются, имеют разные значения разбиения в узле на пути к этим листьям. 
Если условие разбиения выполнено, объект переходит в левое поддерево, в противном случае - в правое.

Если модель использует комбинацию некоторых входных признаков вместо того, чтобы использовать их по отдельности, 
вычисляется и выводится средняя значимость этих признаков. 

Сначала рассчитывается важность признака для комбинации этих признаков. 
Затем полученное значение делится на количество признаков в комбинации и присваивается каждому из признаков.
"""
```

[+1] Сколько суммарно листьев у depth-wise tree глубины D? (обозначим ответ за L)

 - Листьями являются те узлы, у которых нет потомков. Посколько у нас строяться полные сбалансированные деревья, 
то количество листев в дереве глубины D будет равно $L = {2}^{{D}}$

```python
L = 2**D
L
```

[+2] Для тренировочной и валидационной части мы сделаем предсказания. Но не в виде прогноза целевой переменной, а в виде перевода в новое признаковое пространство. 

Каждый объект, когда прогоняется через бустинг, попадает в конкретный лист в каждом дереве. Если в каждом дереве L листьев, тогда всего у бустинга (N * L) листьев. Таким образом, объект можно представить как вектор из N индексов, где на i-й позиции номер листа, в который попал объект в i-ом дереве.

**[+2]** Для тренировочной и валидационной части мы сделаем предсказания. Но не в виде прогноза целевой переменной, а в виде перевода в новое признаковое пространство. 

Каждый объект, когда прогоняется через бустинг, попадает в конкретный лист в каждом дереве. Если в каждом дереве L листьев, тогда всего у бустинга (N * L) листьев. Таким образом, объект можно представить как вектор из N индексов, где на i-й позиции номер листа, в который попал объект в i-ом дереве.

```python
leaf_indexs_train = gbr.calc_leaf_indexes(X_train, ntree_start=0, thread_count=-1, verbose=False)
leaf_indexs_test = gbr.calc_leaf_indexes(X_test, ntree_start=0, thread_count=-1, verbose=False)
```

```python
leaf_indexs_train.shape, leaf_indexs_test.shape
```

**[+2]** Перевести индексы листьев из пространства `[0...L)` (индексы уникальны в рамках дерева) в формат `[0...N*L)` (индексы уникальны в рамках всего ансамбля). 

- допустим мы имеем 3 дерева, объект попал во 2-й лист в первом дереве, в 5-й во втором и в 0-й в третьем.
- на выходе получаем вектор **[2, 5, 0]** (индексы - локальные по каждому дереву)
- его необходимо перевести в **[2, 1*L+5, 2*L+0]** (индексы - глобальные по всему бустингу)
- пример результата на диаграмме

Вывести первые 10 строк и 10 стобцов старой и новой матрицы индексов
(если использована кросс-валидация, по каждому фолду)
+ размерности старой и новой матрицы индексов
+ комментарием в ячейке полученные L и N*L

```python
ensemble_indexs_train = np.array([np.arange(N) * L] * leaf_indexs_train.shape[0]) + leaf_indexs_train
ensemble_indexs_test = np.array([np.arange(N) * L] * leaf_indexs_test.shape[0]) + leaf_indexs_test
```

```python
leaf_indexs_train[:10, :10], leaf_indexs_test[:10, :10]
```

```python
ensemble_indexs_train[:10, :10], ensemble_indexs_test[:10, :10]
```

```python
leaf_indexs_train.shape, leaf_indexs_test.shape
```

```python
ensemble_indexs_train.shape, ensemble_indexs_test.shape
```

```python
# 16, 3200: L, L * N
```

**[+2]** предыдущее задание должно быть выполнено с помощью матричной алгебры numpy

**One-Hot Encoding**

**[+2]** Имея глобальные индексы листьев, можем перевести их в пространство нулей и единиц:

- пусть на предыдущем шаге мы получили индексы [1, 4, 7), а всего индексов 8
- перевести их в формат [0, 1, 0, 0, 1, 0, 0, 1]

Это и будет новым признаковым пространством, в которое мы перевели датасет с помощью градиентного бустинга

```python
ohe_train = np.ndarray(shape=(ensemble_indexs_train.shape[0], N * L), 
                                buffer=np.zeros(N * L * ensemble_indexs_train.shape[0],
                                                dtype=np.int), dtype=np.int)
```

```python
ohe_test = np.ndarray(shape=(ensemble_indexs_test.shape[0], N * L), 
                                buffer=np.zeros(N * L * ensemble_indexs_test.shape[0],
                                                dtype=np.int), dtype=np.int)
```

```python
ohe_train.shape, ohe_test.shape
```

```python
for obj, indx in zip(ohe_train, ensemble_indexs_train):
    obj[indx] = 1
```

```python
for obj, indx in zip(ohe_test, ensemble_indexs_test):
    obj[indx] = 1
```

**Confidence Bound**

- **[+2]** На веса модели (исключая bias) накинуть случайный шум 10000 раз.
например, $w' = w z,\ r \sim \mathcal{N}(1,\ 0.2)$
.
- **[+1]** с помощью каждого инстанса зашумлённой модели делаем предикт

```python
w = lassolinreg.coef_
```

```python
rnd = np.random.RandomState(19)
for i in range(1000):
    z = rnd.normal(1, np.sqrt(0.2), len(lassolinreg.coef_) - 1)
    ws = w[1:] * z
    w[1:] = ws
w
```

```python
lassolinreg.coef_[1:] = w[1:]
```

- **[+1]** *какие квантили отвечают за* нижнюю и верхнюю границу 95%-го доверительного интервала? (LCB и UCB)

 - LCB - Lower confidence bounds = 0.025
 - UCB - Upper confidence bounds = 0.975

```python
M = 100
a = 5
b = 6
x = X_test['date'][M * a:M * b]
y = y_test.values[M * a:M * b]

a = 1.0 * np.array(y)
n = len(y) - 1
m, se = np.mean(y), scipy.stats.sem(y)
ci = se * scipy.stats.t.ppf((1 + 0.95) / 2., n)
```

- **[+1]** взять данные LCB и UCB оценку по каждому объекту (таймстемпу) валидационной выборки

```python
LCB = y - ci
UCB = y + ci
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y - ci, y + ci, color='b', alpha=.1)
```
